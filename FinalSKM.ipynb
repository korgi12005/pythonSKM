```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
from scipy import linalg
```

```python
x= np.array([0,28,62,118,155,210,272,360])
y= np.array([1.00,0.69,0.51,0.23,0.14,0.067,0.041,0.018])
plt.plot(x,y,'o',color='tab:orange')
pass
```

```python
lgY=np.log(y)
lgY
```

```python
m = len(x)
X = np.stack((np.ones(m), x)).T
beta=linalg.solve(X.T@X, X.T@lgY)
```

```python
beta0, beta1=beta
beta0, beta1
beta0=np.exp(beta0)
```

```python
plt.plot(x,beta0*np.exp(x*beta1))
plt.plot(x,y,'o')
pass
```

```python
alpha1,alpha2=np.polyfit(x,beta0*np.exp(x*beta1), 1)
alpha2=np.exp(alpha2)
alpha1,alpha2
```

```python
plt.plot(x,alpha2*np.exp(x*alpha1))
plt.plot(x,y,'o')
pass
```
